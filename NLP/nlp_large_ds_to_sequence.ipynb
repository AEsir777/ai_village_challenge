```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
path = tf.keras.utils.get_file('reviews.csv', 
                               'https://drive.google.com/uc?id=13ySLC_ue6Umt9RJYSeM2t-V0kCv-4C-P')
print (path)
```

```python
ds = pd.read_csv(path)
ds.head()
```

```python
sentences = ds['text']
sentences = np.array(sentences)
```

```python
tokenizer = tf.keras.layers.TextVectorization()
tokenizer.adapt(sentences)
```

```python
vocabulary = tokenizer.get_vocabulary(include_special_tokens=True)
vocabulary
```

```python
np.array(vocabulary).shape
```

```python
sequences = tokenizer(sentences)
sequences
```

```python
padded = tf.keras.utils.pad_sequences(sequences, padding="post")
padded.shape
```

```python
truncated = tf.keras.utils.pad_sequences(sequences, truncating="post", maxlen=100)
truncated.shape
```
